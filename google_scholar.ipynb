```python
import matplotlib.pyplot as plt

# number of papers about "Physics-informed machine learning" published 
# per year according to google scholar as of november 2025
nr_of_papers = {"2016": 606,
                "2017": 747,
                "2018": 1350,
                "2019": 2720,
                "2020": 5350,
                "2021": 10300,
                "2022": 18600,
                "2023": 18900,
                "2024": 48600,
                "2025": 18400}

# plot
plt.figure()
plt.bar(nr_of_papers.keys(), nr_of_papers.values())
plt.xlabel("Year")
plt.ylabel("Number of papers")
plt.title('"Physics-informed machine learning" papers published on google scholar')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```
